 **Q-Learning code to an Atari game environment - Tetris**

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from pathlib import Path
from typing import NamedTuple, List
from ale_py import ALEInterface

# Hyperparameters
class Params(NamedTuple):
    total_episodes: int
    learning_rate: float
    gamma: float
    epsilon: float
    n_runs: int
    savefig_folder: Path

params = Params(
    total_episodes=200,
    learning_rate=0.1,
    gamma=0.99,
    epsilon=0.1,
    n_runs=5,
    savefig_folder=Path("./atari_figures/")
)

params.savefig_folder.mkdir(parents=True, exist_ok=True)

# State preprocessing function for Atari environment
def preprocess_state(state):
    # Convert RGB frames to grayscale
    gray = np.mean(state, axis=2).astype(np.uint8)
    # Downsample the frame to reduce dimensions (simpler representation)
    resized_state = gray[::2, ::2]  # Downsample by selecting every second pixel
    return resized_state

class Qlearning:
    def __init__(self, learning_rate, gamma, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate
        self.gamma = gamma
        self.reset_qtable()

    def update(self, state, action, reward, new_state):
        delta = (
            reward
            + self.gamma * np.max(self.qtable[new_state, :])
            - self.qtable[state, action]
        )
        q_update = self.qtable[state, action] + self.learning_rate * delta
        return q_update

    def reset_qtable(self):
        self.qtable = np.zeros((256, 4))  # 256 possible grayscale states, 4 actions

class EpsilonGreedy:
    def __init__(self, epsilon, learner):
        self.epsilon = epsilon
        self.learner = learner  # Pass the learner instance

    def choose_action(self, action_space):
        if np.random.rand() < self.epsilon:
            return np.random.choice(action_space)
        else:
            return np.argmax(self.learner.qtable)
```

```python
def run_atari(params):
    total_rewards = np.zeros(params.total_episodes)
    episode_lengths = np.zeros(params.total_episodes)
    discounted_cumulative_rewards = np.zeros(params.total_episodes)
    exploration_counts = np.zeros(params.total_episodes)

    ale = ALEInterface()
    rom_path = '/usr/local/share/atari/roms/tetris.bin'
    ale.loadROM(rom_path)

    learner = Qlearning(params.learning_rate, params.gamma, 256, 4)
    explorer = EpsilonGreedy(params.epsilon, learner)  # Pass learner here

    for episode in tqdm(range(params.total_episodes)):
        ale.reset_game()
        screen_obs = ale.getScreenRGB()
        state = preprocess_state(screen_obs)

        episode_reward = 0
        episode_length = 0
        episode_discounted_reward = 0
        exploration_count = 0
        discount = 1

        while True:
            is_exploration = np.random.rand() < explorer.epsilon
            action = explorer.choose_action(range(4))

            if is_exploration:
                exploration_count += 1

            reward = ale.act(action)
            screen_obs = ale.getScreenRGB()

            next_state = preprocess_state(screen_obs)
            updated_q_value = learner.update(state, action, reward, next_state)

            learner.qtable[state, action] = updated_q_value
            episode_reward += reward
            episode_discounted_reward += discount * reward

            state = next_state
            episode_length += 1
            discount *= params.gamma

            if ale.game_over():
                break

        total_rewards[episode] = episode_reward
        episode_lengths[episode] = episode_length
        discounted_cumulative_rewards[episode] = episode_discounted_reward
        exploration_counts[episode] = exploration_count

    avg_reward = np.mean(total_rewards)
    avg_episode_length = np.mean(episode_lengths)

    print(f"Average Reward per Episode: {avg_reward}")
    print(f"Average Episode Length: {avg_episode_length}")
    print(f"Average Discounted Reward per Episode: {np.mean(discounted_cumulative_rewards)}")
    print(f"Average Exploration Count per Episode: {np.mean(exploration_counts)}")

    return {
        'total_rewards': total_rewards,
        'episode_lengths': episode_lengths,
        'discounted_rewards': discounted_cumulative_rewards,
        'exploration_counts': exploration_counts
    }

def plot_results(rewards, steps, episodes, discounted_rewards, exploration_counts, params):
    def plot_rewards(rewards, episodes, params):
        cum_rewards = np.cumsum(rewards)

        plt.figure(figsize=(10, 6))
        plt.plot(episodes, cum_rewards, label="Cumulative rewards", color="blue")
        plt.title("Cumulative Rewards over Episodes")
        plt.xlabel("Episodes")
        plt.ylabel("Cumulative Rewards")
        plt.legend()
        plt.grid(True)
        plt.savefig(params.savefig_folder / "cumulative_rewards.png")
        plt.show()

    def plot_steps(steps, episodes, params):
        mean_steps = steps / episodes

        plt.figure(figsize=(10, 6))
        plt.plot(episodes, mean_steps, label="Mean steps", color="green")
        plt.title("Mean Steps per Episode")
        plt.xlabel("Episodes")
        plt.ylabel("Mean Steps")
        plt.legend()
        plt.grid(True)
        plt.savefig(params.savefig_folder / "mean_steps.png")
        plt.show()

    def plot_discounted_rewards(discounted_rewards, episodes, params):
        plt.figure(figsize=(10, 6))
        plt.plot(episodes, discounted_rewards, label="Discounted Rewards", color="orange")
        plt.title("Discounted Cumulative Rewards over Episodes")
        plt.xlabel("Episodes")
        plt.ylabel("Discounted Rewards")
        plt.legend()
        plt.grid(True)
        plt.savefig(params.savefig_folder / "discounted_rewards.png")
        plt.show()

    def plot_exploration_counts(exploration_counts, episodes, params):
        plt.figure(figsize=(10, 6))
        plt.plot(episodes, exploration_counts, label="Exploration Counts", color="red")
        plt.title("Exploration Counts over Episodes")
        plt.xlabel("Episodes")
        plt.ylabel("Exploration Count")
        plt.legend()
        plt.grid(True)
        plt.savefig(params.savefig_folder / "exploration_counts.png")
        plt.show()

    plot_rewards(rewards, episodes, params)
    plot_steps(steps, episodes, params)
    plot_discounted_rewards(discounted_rewards, episodes, params)
    plot_exploration_counts(exploration_counts, episodes, params)

metrics = run_atari(params)
plot_results(
    metrics['total_rewards'],
    metrics['episode_lengths'],
    np.arange(1, params.total_episodes + 1),
    metrics['discounted_rewards'],
    metrics['exploration_counts'],
    params
)
```
